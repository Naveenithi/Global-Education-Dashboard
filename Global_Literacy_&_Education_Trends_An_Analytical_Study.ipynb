# Step 1: Load → Clean → Save All 5 Datasets

```
import pandas as pd

# Required header (prevents blocking)
storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'}

# -------------------------------
# 1. LOAD DATASETS FROM URLs
# -------------------------------

adult_url = "https://ourworldindata.org/grapher/literacy-rate-adults.csv?v=1&csvType=full&useColumnShortNames=true"
youth_url = "https://ourworldindata.org/grapher/literacy-rate-of-young-men-and-women.csv?v=1&csvType=full&useColumnShortNames=true"
illiterate_url = "https://ourworldindata.org/grapher/literate-and-illiterate-world-population.csv?v=1&csvType=full&useColumnShortNames=true"
gdp_url = "https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true"
schooling_url = "https://ourworldindata.org/grapher/literacy-rates-vs-average-years-of-schooling.csv?v=1&csvType=full&useColumnShortNames=true"

adult_df = pd.read_csv(adult_url, storage_options=storage_options)
youth_df = pd.read_csv(youth_url, storage_options=storage_options)
illiterate_df = pd.read_csv(illiterate_url, storage_options=storage_options)
gdp_df = pd.read_csv(gdp_url, storage_options=storage_options)
schooling_df = pd.read_csv(schooling_url, storage_options=storage_options)

# -------------------------------
# 2. STANDARDIZE COMMON COLUMNS
# -------------------------------

def standardize_common(df):
    return df.rename(columns={
        'Entity': 'country',
        'Code': 'country_code',
        'Year': 'year'
    })

adult_df = standardize_common(adult_df)
youth_df = standardize_common(youth_df)
illiterate_df = standardize_common(illiterate_df)
gdp_df = standardize_common(gdp_df)
schooling_df = standardize_common(schooling_df)

# -------------------------------
# 3. RENAME VALUE COLUMNS
# -------------------------------

# Adult Literacy
adult_df = adult_df.rename(columns={
    'Literacy rate, adult total (% of people ages 15 and above)': 'adult_literacy_rate'
})

# Youth Literacy
youth_df = youth_df.rename(columns={
    'Literacy rate, youth male (% of males ages 15-24)': 'youth_literacy_male',
    'Literacy rate, youth female (% of females ages 15-24)': 'youth_literacy_female'
})

# Illiterate Population
illiterate_df = illiterate_df.rename(columns={
    'Illiterate population, total': 'illiterate_total',
    'Illiterate population, male': 'illiterate_male',
    'Illiterate population, female': 'illiterate_female'
})

# GDP
gdp_df = gdp_df.rename(columns={
    'GDP per capita, PPP (constant 2017 international $)': 'gdp_per_capita'
})

# Schooling
schooling_df = schooling_df.rename(columns={
    'Average years of schooling': 'avg_years_schooling'
})

# -------------------------------
# 4. FINAL COLUMN CLEANUP (SAFE)
# -------------------------------

def clean_column_names(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
    return df

adult_df = clean_column_names(adult_df)
youth_df = clean_column_names(youth_df)
illiterate_df = clean_column_names(illiterate_df)
gdp_df = clean_column_names(gdp_df)
schooling_df = clean_column_names(schooling_df)

# -------------------------------
# 5. VERIFY STRUCTURE
# -------------------------------

print("Adult:", adult_df.columns)
print("Youth:", youth_df.columns)
print("Illiterate:", illiterate_df.columns)
print("GDP:", gdp_df.columns)
print("Schooling:", schooling_df.columns)

# -------------------------------
# 6. SAVE CLEAN CSV FILES
# -------------------------------

adult_df.to_csv("adult_literacy_cleaned.csv", index=False)
youth_df.to_csv("youth_literacy_cleaned.csv", index=False)
illiterate_df.to_csv("illiterate_population_cleaned.csv", index=False)
gdp_df.to_csv("gdp_per_capita_cleaned.csv", index=False)
schooling_df.to_csv("schooling_cleaned.csv", index=False)

print("\n✅ All datasets cleaned, renamed, and saved correctly!")
```

# Step 2: Load Cleaned CSVs + Create Required DataFrames

# 1. Load Cleaned CSV Files

```
import pandas as pd

adult_df = pd.read_csv("adult_literacy_cleaned.csv")
youth_df = pd.read_csv("youth_literacy_cleaned.csv")
illiterate_df = pd.read_csv("illiterate_population_cleaned.csv")
gdp_df = pd.read_csv("gdp_per_capita_cleaned.csv")
schooling_df = pd.read_csv("schooling_cleaned.csv")
```

# 2. Create df_literacy (Adult + Youth Literacy)

# Merge Literacy

```
df_literacy = pd.merge(
    adult_df,
    youth_df,
    on=['country', 'country_code', 'year'],
    how='inner'
)
```

# Illiteracy

```
df_illiteracy = illiterate_df.copy()
```

# GDP + Schooling

```
df_gdp_schooling = pd.merge(
    gdp_df,
    schooling_df,
    on=['country', 'country_code', 'year'],
    how='inner'
)
```

# Step 3: Data Cleaning

```
df_literacy
df_illiteracy
df_gdp_schooling
```

# 1. Handle Missing Values

Check Missing Values

```
print("Literacy Missing:\n", df_literacy.isnull().sum())
print("\nIlliteracy Missing:\n", df_illiteracy.isnull().sum())
print("\nGDP & Schooling Missing:\n", df_gdp_schooling.isnull().sum())
```

Drop Missing

```
df_literacy = df_literacy.dropna()
df_illiteracy = df_illiteracy.dropna()
df_gdp_schooling = df_gdp_schooling.dropna()

# Rows with missing values were removed to maintain data integrity
```

# 2. Remove Duplicates

```
df_literacy = df_literacy.drop_duplicates()
df_illiteracy = df_illiteracy.drop_duplicates()
df_gdp_schooling = df_gdp_schooling.drop_duplicates()
```

# 3. Standardize Country Names

```
country_mapping = {
    "United States": "USA",
    "United Kingdom": "UK"
}

for df in [df_literacy, df_illiteracy, df_gdp_schooling]:
    df['country'] = df['country'].replace(country_mapping)
```

# 4. Understand Shape & Structure

```
print("Literacy Shape:", df_literacy.shape)
print("Illiteracy Shape:", df_illiteracy.shape)
print("GDP & Schooling Shape:", df_gdp_schooling.shape)
```

```
df_literacy.info()
df_illiteracy.info()
df_gdp_schooling.info()
```

# 5. Identify Unusual Values

```
df_literacy.describe()
df_illiteracy.describe()
df_gdp_schooling.describe()
```

Check Invalid Values

```
# Literacy should be between 0 and 100
df_literacy = df_literacy[
    (df_literacy['adult_literacy_rate'] <= 100) &
    (df_literacy['youth_literacy_male'] <= 100) &
    (df_literacy['youth_literacy_female'] <= 100)
]
```

# 6. Filter Years (1990–2023)

```
df_literacy = df_literacy[
    (df_literacy['year'] >= 1990) & (df_literacy['year'] <= 2023)
]

df_illiteracy = df_illiteracy[
    (df_illiteracy['year'] >= 1990) & (df_illiteracy['year'] <= 2023)
]

df_gdp_schooling = df_gdp_schooling[
    (df_gdp_schooling['year'] >= 1990) & (df_gdp_schooling['year'] <= 2023)
]
```

# 7. Remove Aggregate Regions

```
exclude = ['World', 'Asia', 'Europe', 'Africa', 'High income', 'Low income']

df_literacy = df_literacy[~df_literacy['country'].isin(exclude)]
df_illiteracy = df_illiteracy[~df_illiteracy['country'].isin(exclude)]
df_gdp_schooling = df_gdp_schooling[~df_gdp_schooling['country'].isin(exclude)]
```

# 8. Final Check

```
print("Final Literacy:", df_literacy.shape)
print("Final Illiteracy:", df_illiteracy.shape)
print("Final GDP & Schooling:", df_gdp_schooling.shape)
```

# 9. Save Cleaned Data

```
df_literacy.to_csv("df_literacy_cleaned_final.csv", index=False)
df_illiteracy.to_csv("df_illiteracy_cleaned_final.csv", index=False)
df_gdp_schooling.to_csv("df_gdp_schooling_cleaned_final.csv", index=False)

print("✅ Cleaned datasets saved successfully!")
```

```
df_literacy['literacy_gender_gap'] = (
    df_literacy['youth_literacy_male'] - df_literacy['youth_literacy_female']
)
```

3. Youth Literacy Average

```
df_literacy['youth_literacy_avg'] = (
    df_literacy['youth_literacy_male'] + df_literacy['youth_literacy_female']
) / 2
```

4. GDP per Schooling Year

```
df_gdp_schooling['gdp_per_schooling_year'] = (
    df_gdp_schooling['gdp_per_capita'] / df_gdp_schooling['avg_years_schooling']
)
```

```
df_gdp_schooling = df_gdp_schooling[
    df_gdp_schooling['avg_years_schooling'] > 0
]
```

# 5. Education Index (Custom Feature)

First merge literacy + schooling:

```
df_education = pd.merge(
    df_literacy,
    df_gdp_schooling,
    on=['country', 'country_code', 'year'],
    how='inner'
)
```

Normalize values (0 to 1 scale):

```
df_education['literacy_norm'] = df_education['adult_literacy_rate'] / 100

df_education['schooling_norm'] = (
    df_education['avg_years_schooling'] / df_education['avg_years_schooling'].max()
)
```

Create Education Index:

```
df_education['education_index'] = (
    df_education['literacy_norm'] + df_education['schooling_norm']
) / 2
```

# 6. Literacy Growth Rate (Year-over-Year)

```
df_literacy = df_literacy.sort_values(by=['country', 'year'])

df_literacy['literacy_growth_rate'] = (
    df_literacy.groupby('country')['adult_literacy_rate']
    .pct_change() * 100
)
```

Final Check

```
print(df_literacy.columns)
print(df_gdp_schooling.columns)
print(df_education.columns)
```

Save Final Feature Data
